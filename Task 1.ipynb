# **Visualizations**

**Scatter plot to analyze relationships between variables.**

Sepal Length vs Sepal Width: This plot shows the relationship between the length and width of the sepals for each flower. The points are colored based on the species column, allowing us to see how the different species cluster based on these two features.(matplotlib.pyplot to set the figure size, add titles and labels to the axes, and display the plots).

```python
sns.scatterplot(x = "sepal_length", y = "sepal_width", hue = "species", data = data, palette= ('red', 'green','blue'))
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length(cm)")
plt.ylabel("Sepal Width(cm)")
plt.show()
```

Petal Length vs Petal Width: Similar to the first plot, this one visualizes the relationship between the length and width of the petals, also colored by species. This plot is often very effective in separating the different iris species. (matplotlib.pyplot to set the figure size, add titles and labels to the axes, and display the plots.)

```python
sns.scatterplot(x = "petal_length", y = "petal_width", hue = "species", data = data, palette= ('pink', 'black','green'))
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length(cm)")
plt.ylabel("Petal Width(cm)")
plt.show()
```

**Histogram to examine data distribution.**

In this cell I generate a histogram using seaborn to visualize the distribution of the sepal_length column from the data DataFrame.

kde=True adds a Kernel Density Estimate line to the histogram, showing the probability density of the data.
color='blue' sets the color of the histogram bars.
plt.title(), plt.xlabel(), and plt.ylabel() are used to add a title and labels to the plot. plt.show() displays the generated plot.

```python
sns.histplot(data['sepal_length'], kde = True, color = 'blue')
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length(cm)")
plt.ylabel("Frequency")
plt.show()
```

**Box plot to detect outliers and spread of values.**

In this cell I generate a box plot using seaborn to visualize the distribution of the sepal_length for each species in the data DataFrame.

x='species' sets the species on the x-axis.
y='sepal_length' sets the sepal length on the y-axis.
data=data specifies the DataFrame to use.
palette is used to set the colors for each box plot.
plt.title(), plt.xlabel(), and plt.ylabel() are used to add a title and labels to the plot.
plt.show() displays the generated plot.

```python
sns.boxplot(x = 'species', y = 'sepal_length', data = data, palette= ('red', 'green','blue'))
plt.title("Box Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length(cm)")
plt.show()
```
